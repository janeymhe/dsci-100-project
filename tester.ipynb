```r
library(tidyverse)
library(repr)
library(tidymodels)


url_players <- "https://drive.google.com/uc?export=download&id=1Mw9vW0hjTJwRWx0bDXiSpYsO3gKogaPz"
url_sessions <- "https://drive.google.com/uc?export=download&id=14O91N5OlVkvdGxXNJUj5jIsV5RexhzbB"

players <- read_csv(url_players)
players

sessions <- read_csv(url_sessions)|>
            separate(start_time, into = c("start_date", "start_time"), sep = " ")|>
            separate(end_time, into = c("end_date", "end_time"), sep = " ")
sessions

options(repr.plot.width=10, repr.plot.height=7)
players_plot <- players |> 
                ggplot(aes(x = age, y = played_hours))+
                geom_point(aes(color = experience, alpha = 0.4))+
                labs(x = "Age of the player",
                     y = "Time played in hours",
                     color = "Minecraft experience level")+
                 ggtitle("Age vs time played in hours")+
                 theme(text = element_text(size = 20))
players_plot

sessions_plot <- sessions |> 
                 ggplot(aes(x = start_time, y = original_start_time))+
                 geom_point(aes(alpha = 0.4))+
                 labs(x = "Start time of unique session",
                      y = "Scheduled start time (UNIX timestamp format)")+
                 ggtitle("Start of unique session vs scheduled start time UNIX timestamp format ")+
                 theme(text = element_text(size = 15))
sessions_plot
```
